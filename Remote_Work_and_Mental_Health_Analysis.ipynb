```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set visualization styles
sns.set(color_codes=True)

# matplotlib inline

```

```python
# Pie chart for gender distribution
gender_counts = RM['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'yellow'])
plt.title('Gender Distribution')
plt.show()







```

```python
# Plotting the age distribution
plt.figure(figsize=(8, 6))
sns.histplot(RM['Age'], bins=20, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Industry distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Industry', data=RM, palette='viridis')
plt.title('Industry Distribution')
plt.xlabel('Industry')
plt.ylabel('Count')
plt.show()
```
